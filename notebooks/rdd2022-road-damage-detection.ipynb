```python
from ultralytics import YOLO
import os
import yaml
import shutil
import zipfile

print("="*50)
print("ğŸš€ ROAD DAMAGE DETECTION - FULL TRAINING")
print("="*50)
```

```python
BASE_PATH = '/kaggle/input/submission'
TRAIN_IMG_PATH = f'{BASE_PATH}/train-20260109T210025Z-3-003/train/images'
TRAIN_LBL_PATH = f'{BASE_PATH}/train-20260109T210025Z-3-003/train/labels'
TEST_IMG_PATH = f'{BASE_PATH}/test-20260109T184817Z-3-001/test/images'

# Verify
print("\nğŸ“‚ Checking dataset...")
train_imgs = len(os.listdir(TRAIN_IMG_PATH))
train_lbls = len(os.listdir(TRAIN_LBL_PATH))
test_imgs = len(os.listdir(TEST_IMG_PATH))

print(f"âœ… Train images: {train_imgs}")
print(f"âœ… Train labels: {train_lbls}")
print(f"âœ… Test images: {test_imgs}")
```

```python
print("\nğŸ“ Creating configuration...")

data_yaml = {
    'path': f'{BASE_PATH}/train-20260109T210025Z-3-003',
    'train': 'train/images',
    'val': 'train/images',
    'nc': 5,
    'names': ['Longitudinal Crack', 'Transverse Crack', 'Alligator Crack', 'Other Corruption', 'Pothole']
}

yaml_path = '/kaggle/working/data.yaml'
with open(yaml_path, 'w') as f:
    yaml.dump(data_yaml, f)

print("âœ… Configuration created!")
```

```python
print("\n Loading YOLOv8 Medium...")
model = YOLO('yolov8m.pt')
print(" Model loaded!")
```

```python
print("\n" + "="*50)
print(" STARTING TRAINING (1-2 hours)")
print("="*50)
print(" Go get coffee, this will take a while...")
print(" Training will show progress bars\n")

results = model.train(
    data=yaml_path,
    epochs=50,
    imgsz=640,
    batch=16,
    name='road_damage_model',
    patience=15,
    save=True,
    device=0,
    workers=2,
    verbose=True,
    plots=True,
    hsv_h=0.015,
    hsv_s=0.7,
    hsv_v=0.4,
    translate=0.1,
    scale=0.5,
    fliplr=0.5,
    mosaic=1.0,
)

print("\n Training complete!")
```

```python
print("\n Evaluating model...")
metrics = model.val()

print(f"\n RESULTS:")
print(f"  mAP50:     {metrics.box.map50:.4f}")
print(f"  mAP50-95:  {metrics.box.map:.4f}")
print(f"  Precision: {metrics.box.mp:.4f}")
print(f"  Recall:    {metrics.box.mr:.4f}")

# Save metrics
with open('/kaggle/working/metrics.txt', 'w') as f:
    f.write(f"mAP50: {metrics.box.map50:.4f}\n")
    f.write(f"mAP50-95: {metrics.box.map:.4f}\n")
    f.write(f"Precision: {metrics.box.mp:.4f}\n")
    f.write(f"Recall: {metrics.box.mr:.4f}\n")
```

```python
print("\n Generating test predictions...")

OUTPUT_DIR = '/kaggle/working/predictions'
os.makedirs(OUTPUT_DIR, exist_ok=True)

test_images = sorted(os.listdir(TEST_IMG_PATH))
total = len(test_images)

for idx, img_name in enumerate(test_images):
    img_path = f'{TEST_IMG_PATH}/{img_name}'
    results_pred = model.predict(img_path, conf=0.25, iou=0.45, verbose=False)
    
    txt_name = img_name.replace('.jpg', '.txt').replace('.png', '.txt')
    txt_path = f'{OUTPUT_DIR}/{txt_name}'
    
    with open(txt_path, 'w') as f:
        for box in results_pred[0].boxes:
            cls = int(box.cls[0])
            conf = float(box.conf[0])
            x, y, w, h = box.xywhn[0].tolist()
            f.write(f"{cls} {x:.6f} {y:.6f} {w:.6f} {h:.6f} {conf:.6f}\n")
    
    if (idx + 1) % 100 == 0 or (idx + 1) == total:
        print(f"  {idx + 1}/{total} done")

print(f" All predictions generated!")
```

```python
print("\n Creating submission.zip...")

zip_path = '/kaggle/working/submission.zip'
with zipfile.ZipFile(zip_path, 'w') as zipf:
    for txt in os.listdir(OUTPUT_DIR):
        zipf.write(f'{OUTPUT_DIR}/{txt}', f'predictions/{txt}')

print(f" Submission created ({len(os.listdir(OUTPUT_DIR))} files)")
```
